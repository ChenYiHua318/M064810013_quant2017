# Q1寫一個函式，抓取 Google Finance 的股價歷史資料

```python
import requests
import pandas as pd
from datetime import datetime
import io

def get_params(symbol,startdate,enddate):
    params = {'q': symbol,
              'startdate': startdate,
              'enddate': enddate,
              'output': "csv"}
    return params
    
def main():
    symbol = input('股票代碼')
    startdate = input('開始日(格式2017-01-01):')
    enddate=datetime
    url = 'http://finance.google.com/finance/historical'
    r = requests.get(url, get_params(symbol,startdate,enddate))
    df = pd.read_csv(io.StringIO(r.text))
    return df
    
main()
```
