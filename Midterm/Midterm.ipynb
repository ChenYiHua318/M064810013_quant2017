# Q1. 畫出 Diamond Shape

```python
import math
```

```python
def reverse_pyramid(n):
    for k in range (1,n-math.floor(n/2)+1):
        print(' '*(n-(k+math.floor(n/2))) , '*' * (2*k-1))
    for y in range(n-math.floor(n/2)-1,-1,-1):
        print(' '*(n-(y+math.floor(n/2))) , '*' * (2*y-1))
    
reverse_pyramid(9)
```

```python
def xmas_tree(n):
    for i in range(1,n):
        for k in range (1,n-math.floor(n/2)+1):
            print(' '*(n-(k+math.floor(n/2))) , '*' * (2*k-1))
xmas_tree(9)
```

```python
import ffn
%matplotlib inline
import numpy as np
import pandas_datareader.data as web
from datetime import datetime
import pickle
import talib
from talib import MA_Type
import matplotlib.pyplot as plt
import kchart
```

```python
#下載股價資料
with open('stockdata', 'rb') as f:
    data = pickle.load(file=f)
```

```python
data
```

```python
#取出Google股價資訊
df = data['GOOG']
```

```python
#Bollinger Bands  
def BBANDS(df, n):
#計算移動平均線、標準差
    MA =pd.Series.rolling(df['Close'],window=n,center=False).mean() 
    MSD = pd.Series.rolling(df['Close'],window=n,center=False).std()
    b0=MA
    B0=pd.Series(b0,name='MA')
    df=df.join(B0)
#計算壓力線
    b1=MA+2*MSD
    B1=pd.Series(b1,name='Bollinger_resistance')
    df=df.join(B1)
#計算支撐線    
    b2=MA+(-2)*MSD
    B2=pd.Series(b2,name='Bollinger_support')
    df=df.join(B2)
#計算布林極限指標%B=(收盤價-支撐線值)/(壓力線值-支撐線值)    
    b3_1=(df['Close']-df['Bollinger_support'])
    b3_2=(df['Bollinger_resistance']-df['Bollinger_support'])
    b3=b3_1/b3_2
    B3=pd.Series(b3,name='Bollinger_%B')
    df=df.join(B3)
    
    return df
```

```python
#設定為20日移動平均，將函式結果輸出為DataFrame
df=pd.DataFrame(BBANDS(df, 20))
df
```

```python
kchart.draw(df)
```

```python
#畫出布林通道，加入收盤價走勢
df[['MA','Bollinger_resistance','Bollinger_support','Close']].plot(legend='best', figsize = (12.5, 10))
```

# Q5. 策略回測

```python
import requests
import pandas as pd
import math
import ffn
%matplotlib inline
import numpy as np
import pandas_datareader.data as web
from datetime import datetime
import pickle
import talib
from talib import MA_Type
import matplotlib.pyplot as plt
```

```python
# 計算 MaxDD
def DrawDownAnalysis(cumRet):
    dd_series = ffn.core.to_drawdown_series(cumRet)
    dd_details = ffn.core.drawdown_details(dd_series)
    return dd_details['drawdown'].min(), dd_details['days'].max()
```

```python
# 利用策略產生的持有部位資訊，計算底下四個指標來判斷投資績效
# sharpe ratio: 判斷報酬的好壞跟穩定度，數值越大越好
# maxdd: maximum drawdown, 最糟糕的狀況會賠幾 %
# maxddd: maximum drawdown duration, 低於上一次最高報酬的天數
# cumRet[-1]: 最後賺的 % 數

def indicators(df):
    dailyRet = df['Close'].pct_change()
    excessRet = (dailyRet - 0.04/252)[df['positions'] == 1]
    SharpeRatio = np.sqrt(252.0)*np.mean(excessRet)/np.std(excessRet)
    
    cumRet = np.cumprod(1+excessRet)
    
    maxdd, maxddd = DrawDownAnalysis(cumRet)
    
    return SharpeRatio, maxdd, maxddd, cumRet[-1]
```

```python
# 這是我們的策略的部分
# 主要只是要算出進出的訊號 signals 跟何時持有部位 positions
# 底下是一個突破系統的範例

def Breakout_strategy(df):
    # Donchian Channel
    df['20d_high'] = np.round(pd.Series.rolling(df['Close'], window=20).max(), 2)
    df['10d_low'] = np.round(pd.Series.rolling(df['Close'], window=10).min(), 2)

    has_position = False
    df['signals'] = 0
    for t in range(2, df['signals'].size):
        if df['Close'][t] > df['20d_high'][t-1]:
            if not has_position:
                df.loc[df.index[t], 'signals'] = 1
                has_position = True
        elif df['Close'][t] < df['10d_low'][t-1]:
            if has_position:
                df.loc[df.index[t], 'signals'] = -1
                has_position = False

    df['positions'] = df['signals'].cumsum().shift()
    return df
```

```python
def RSI_7030_strategy(df):
    df['RSI'] = talib.RSI(df['Close'].values)
    
    has_position = False
    df['signals'] = 0
    for t in range(2, df['signals'].size):
        if df['RSI'][t-1] < 30:
            if not has_position:
                df.loc[df.index[t], 'signals'] = 1
                has_position = True
        elif df['RSI'][t-1] > 70:
            if has_position:
                df.loc[df.index[t], 'signals'] = -1
                has_position = False

    df['positions'] = df['signals'].cumsum().shift()
    return df
```

```python
def BBands_strategy(df):
    df['UBB'], df['MBB'], df['LBB'] = talib.BBANDS(df['Close'].values, matype=MA_Type.T3)

    has_position = False
    df['signals'] = 0
    for t in range(2, df['signals'].size):
        if df['Close'][t] < df['LBB'][t-1]:
            if not has_position:
                df.loc[df.index[t], 'signals'] = 1
                has_position = True
        elif df['Close'][t] > df['UBB'][t-1]:
            if has_position:
                df.loc[df.index[t], 'signals'] = -1
                has_position = False

    df['positions'] = df['signals'].cumsum().shift()
    return df
```

```python
############################################ 我的策略 ##############################################

#計算MA10、MA20、Martin(流動性衡量:Martin愈小、流動性愈好)
D1_1=pd.Series.rolling(df['Close'],window=10,center=False).mean()
D1=pd.Series(D1_1,name='MA_10day')
df=df.join(D1)

D2_1=pd.Series.rolling(df['Close'],window=20,center=False).mean()
D2=pd.Series(D2_1,name='MA_20day')
df=df.join(D2)

D3_1=df['Close']-df['Open']
D3=pd.Series(D3_1,name='PriceChange')
df=df.join(D3)

D4_1=((df['PriceChange'])*(df['PriceChange']))/df['Volume']
D4=pd.Series(D4_1,name='Martin')
df=df.join(D4)
df
```

```python
############################################ 我的策略 ##############################################

def M064810013_strategy(df):
    
    #設置進出場點
    
    has_position = False
    df['signals'] = 0
    for t in range(2, df['signals'].size):
        #若布林極限指標突破0.2 or 突破0.8 + 短天期突破長天期均線 + 流動性指標Martin<0 (流動性好)時-----進場
        if 0.2>df['Bollinger_%B'][t]>0>df['Bollinger_%B'][t-1] and 1>df['Bollinger_%B'][t]>0.8>df['Bollinger_%B'][t-1]and df['MA_10day'][t]>df['MA_20day'][t] and df['MA_10day'][t-1]<df['MA_20day'][t-1] and df['Martin'][t]<0 :           
            if not has_position:
                df.loc[df.index[t], 'signals'] = 1
                has_position = True
        #若布林極限指標跌破0.8 or 跌破0.5 + 短天期跌破長天期均線 + 流動性指標Martin>0 (流動性不好)時---出場        
        elif df['Bollinger_%B'][t]<0.8<df['Bollinger_%B'][t-1] and df['Bollinger_%B'][t]<0.5<df['Bollinger_%B'][t-1] and df['MA_10day'][t]<df['MA_20day'][t] and df['MA_10day'][t-1]>df['MA_20day'][t-1] and df['Martin'][t]>0 :
                
            if has_position:
                df.loc[df.index[t], 'signals'] = -1
                has_position = False

    df['positions'] = df['signals'].cumsum().shift()
    return df
```

```python
def apply_strategy(strategy, df):
    return strategy(df)
```

```python
# 計算各支股票的回測結果
results = []

strategies = [Breakout_strategy, RSI_7030_strategy, BBands_strategy, M064810013_strategy]

for symbol in data:
    for strategy in strategies:
        try:
            apply_strategy(strategy, data[symbol])
            if np.all(data[symbol]['signals']==0):
                print("Symbol:", symbol, "使用", strategy.__name__, "策略沒有出現買賣訊號。")
                continue
            SharpeRatio, maxdd, maxddd, finalRet = indicators(data[symbol])
            days = (data[symbol].index[-1] - data[symbol].index[0]).days
            results.append((SharpeRatio, maxdd, maxddd, finalRet, days,
                            data[symbol][data[symbol]['signals'] > 0]['signals'].sum(), symbol, strategy.__name__))
        except Exception as e:
            print("Error occurs at symbol:", symbol, "Strategy:", strategy.__name__, "==>", e.args)
```

```python
results_df = pd.DataFrame(results, columns=['sharpe','MaxDrawDown','MaxDrawDownDuration','returns', 
                                            'days', 'entries','symbol','strategy'])
```
